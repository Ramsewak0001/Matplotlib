```python
Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
Note: Also add title, xlabel, and ylabel to the plot.
```

```python
Scatter Plot:
            A scatter plot represents individual data points as markers on a two-dimensional plane.
            It is useful for visualizing the relationship between two continuous variables and identifying any patterns or correlations.
```

```python
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,y)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Scatter plot')
```

```python
Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
Use the following data:
import numpy as np
For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])
For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])
For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])
For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])
```

```python
subplot() function in Matplotlib is used to create a grid of subplots within a single figure. 
It allows you to divide the figure into multiple smaller plots, each with its own axes. 
This function is particularly useful when you want to display multiple plots side by side or in a grid layout.
```

```python
x = np.array([0, 1, 2, 3, 4, 5])

# Line 1
y1 = np.array([0, 100, 200, 300, 400, 500])
plt.subplot(2, 2, 1)  # 2 rows, 2 columns, plot 1
plt.plot(x, y1)
plt.title("Line 1")

# Line 2
y2 = np.array([50, 20, 40, 20, 60, 70])
plt.subplot(2, 2, 2)  # 2 rows, 2 columns, plot 2
plt.plot(x, y2)
plt.title("Line 2")

# Line 3
y3 = np.array([10, 20, 30, 40, 50, 60])
plt.subplot(2, 2, 3)  # 2 rows, 2 columns, plot 3
plt.plot(x, y3)
plt.title("Line 3")

# Line 4
y4 = np.array([200, 350, 250, 550, 450, 150])
plt.subplot(2, 2, 4)  # 2 rows, 2 columns, plot 4
plt.plot(x, y4)
plt.title("Line 4")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([0, 1, 2, 3, 4, 5])

# Line 1
y1 = np.array([0, 100, 200, 300, 400, 500])
plt.plot(x, y1, c='red')

# Line 2
y2 = np.array([50, 20, 40, 20, 60, 70])
plt.plot(x, y2, c= 'black')

# Line 3
y3 = np.array([10, 20, 30, 40, 50, 60])
plt.plot(x, y3, c='blue')

# Line 4
y4 = np.array([200, 350, 250, 550, 450, 150])
plt.plot(x, y4, c='green')

plt.title('Four Line Plots')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

```python
Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

```python
Bar Plot: 
         A bar plot (or bar chart) displays categorical data using rectangular bars. Each bar represents a category, 
         and the height of the bar indicates the value associated with that category. Bar plots are often used to compare
         different categories or display frequency counts.
        
Use of Bar plot :
         1. Bar plots are used to visualize and compare categorical data.
         2. They provide a clear representation of data across different categories or groups.
         3. Bar plots allow for easy comparison of values or frequencies between categories.
         4. They help identify patterns, trends, and disparities in data.
```

```python
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

```python
import matplotlib.pyplot as plt
plt.bar(company,profit)
plt.xlabel("Company Name")
plt.ylabel("Profit")
plt.title('Bar Plot')
plt.grid()
plt.show()
```

```python
# Use barh for plot hirizontaly
import matplotlib.pyplot as plt
plt.barh(company,profit)
plt.xlabel("Company Name")
plt.ylabel("Profit")
plt.title('Bar Plot')
plt.grid()
plt.show()
```

```python
Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
    box1 = np.random.normal(100, 10, 200)
    box2 = np.random.normal(90, 20, 200)
```

```python
Box plot :
          also known as a box-and-whisker plot, is a statistical visualization tool that provides a concise summary of
          the distribution of a dataset. It displays the quartiles, median, and potential outliers of the data.
            
Box plots are commonly used to analyze and compare the central tendency, spread, and skewness of data across 
different categories or groups.
```

```python
import numpy as np
import matplotlib.pyplot as plt

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

plt.boxplot([box1,box2])
plt.xticks([1,2],["Box 1"," Box 2"])
plt.ylabel("Values")
plt.title("Box Plot")
plt.show()
```
